```python
import numpy as np
import pandas as pd

import yfinance as yf
import datetime
from datetime import date , timedelta
today = date.today()
```

How we can get the latest stock price data

```python
d1 = today.strftime("%Y-%m-%d")
```

```python
end_date = d1
d2 = date.today() - timedelta(days=360)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
d1
```

```python
d2
```

```python
data = yf.download('AAPL' , start= start_date , end = end_date , progress = False)
```

```python
data.head()
```

```python
data = yf.download('BTC-USD' , start = start_date , end = end_date , progress = False)
```

```python
data['Date'] = data.index
```

```python
data = data[['Date' , 'Open' , 'High' , 'Low' , 'Close' , 'Adj Close' , 'Volume']]
```

```python
data.reset_index(drop= True , inplace = True)
```

```python
data.head()
```

```python
# Now let's  visualize the change in bitcoin prices till today
```

```python
import plotly.graph_objects as go
figure = go.Figure(data = [go.Candlestick(x = data['Date'] ,
                                          open = data['Open'],
                                          high = data['High'],
                                          low = data['Low'],
                                          close = data['Close'])])

figure.update_layout(title = "Bitcoin Price Analysis" ,
                     xaxis_rangeslider_visible = False)
```

Now let's have a look at the correlation of all the columns in the data concerning the Close column .

```python
correlation = data.corr()
print(correlation['Close'].sort_values(ascending = False))
```

Now make predictions

```python
# Now I will be using AutoTS library to predict the bitcoin prices for the next 30 days .
```

```python
from autots import AutoTS
```

```python
model = AutoTS(forecast_length = 30 , frequency = 'infer' , ensemble = 'simple')
model = model.fit(data , date_col = 'Date' , value_col = 'Close' , id_col = None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
